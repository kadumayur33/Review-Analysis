## Objective
Analyze synthetic e-commerce product reviews using Pandas.

```python
import pandas as pd
import random

random.seed(42)

categories = ['Electronics', 'Clothing', 'Books', 'Sports']
cities = ['Mumbai', 'Delhi', 'Bangalore', 'Chennai']

data = []
for i in range(100):
    data.append({
        'product': random.choice(['Laptop', 'T-Shirt', 'Novel', 'Shoes', 'Phone']),
        'category': random.choice(categories),
        'rating': random.randint(1, 5),
        'price': random.randint(500, 20000),
        'city': random.choice(cities)
    })

df = pd.DataFrame(data)
df.head()
```

## Task 1: Preview and Explore

```python
df.head(10)
df.tail(5)
df.shape
df.info()
df.describe()
```

### Single vs Multiple Columns

```python
df['rating']
df[['rating', 'price']]
```

## Task 2: Filtering

```python
# Bangalore reviews
bangalore_df = df[df['city'] == 'Bangalore']
bangalore_df.head()

# 5-star and price > 10000
df[(df['rating'] == 5) & (df['price'] > 10000)]

# Electronics or Books with rating < 3
df[((df['category'] == 'Electronics') | (df['category'] == 'Books')) & (df['rating'] < 3)]
```

## Task 3: Value Counts and Visualization

```python
df['category'].value_counts()
df['city'].value_counts()

df['category'].value_counts().plot(kind='bar')
df['rating'].plot(kind='hist', bins=5)
```

## Insights
- Categories with highest counts show demand.
- Rating histogram shows customer satisfaction trend.
